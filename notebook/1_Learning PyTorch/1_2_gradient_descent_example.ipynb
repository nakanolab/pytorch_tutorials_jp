```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
def get_loss(w):
    return w**4 - 16 * w**2 - 12 * w + 96

ws = np.linspace(-5, 5, 101)
plt.plot(ws, get_loss(ws))
```

```python
w = torch.tensor(0.0, requires_grad=True)
loss = get_loss(w)
print(loss)
```

```python
print(w.grad)
loss.backward()
print(w.grad)
```

```python
def iterate(w_init, alpha, n):
    w = torch.tensor(w_init, requires_grad=True)
    w_history = [w_init]
    for _ in range(n):
        loss = get_loss(w)
        loss.backward()
        with torch.no_grad():
            w -= alpha * w.grad
            w.grad.zero_()
        w_history.append(w.item())
    return w_history
```

```python
iterate(0.0, 0.01, 20)
```

```python
iterate(0.0, 0.1, 20)
```

```python
def plot_iterations(w_history):
    ws = np.linspace(-4, 4, 101)
    fig, axes = plt.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios': [2, 1]})
    axes[0].plot(ws, get_loss(ws))
    axes[1].plot(w_history, range(len(w_history)), 'bo-')
    axes[1].invert_yaxis()
```

```python
plot_iterations(iterate(0.0, 0.01, 20))
```

```python
plot_iterations(iterate(0.0, 0.05, 20))
```

```python
plot_iterations(iterate(0.0, 0.03, 20))
```
